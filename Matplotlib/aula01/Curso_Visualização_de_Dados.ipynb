# Scatterplot

```
#Método para plotar um gráfico de dispersão (ou gráfico de pontos)
plt.scatter(x1, y1);
```

É possível plotar um gráfico de linhas em cima de um scatterplot, ligando os pontos correspondentes.

```
#Devem receber os mesmos parâmetros.
plt.scatter(x1, y1, label="Meus pontos", color="red")
plt.legend()
#Propriedade color, para alterar a cor correspondente ao gráfico
plt.plot(x1, y1, color="black")
plt.show()
```

# Configurações Visuais para o Scatterplot





Por padrão, o scatter plot é implementado com a cor azul, o marcador ponto e o tipo de linha sólida:

color = b / 'blue'

marker = '.' / 'point'

linestyle = '-' / 'solid'

```
#Padrão do Scatter
plt.scatter(x1, y1, label="Meus pontos")
plt.legend()
plt.plot(x1, y1)
plt.show()
```

Alterar Propriedades

```
plt.scatter(x1, y1, label="Meus pontos", color="yellow", marker="*")
plt.legend()
plt.plot(x1, y1, color="black")
plt.show()
```

```
plt.scatter(x1, y1, label="Meus pontos", color="blue", marker="^")
plt.legend()
plt.plot(x1, y1, color="black")
plt.show()
```

```
#s é referente ao tamanho do marcador.
plt.scatter(x1, y1, label="Meus pontos", color="cyan", marker="h", s=300)
plt.legend()
plt.plot(x1, y1, color="black", linestyle=":")
plt.show()
```

**Alterando a propriedade color utilizando o formato hexadecimal para as cores**

```
#00 00 00 --> Nível de Vermelho / Nível de Green / Nível de Blue
plt.scatter(x1, y1, label="Meus pontos", color="#000000", marker=".", s=400)
plt.legend()
plt.plot(x1, y1, color="#F08080", linestyle="-")
plt.show()
```

**Salvando Figuras**

Formato PNG

```
plt.scatter(x1, y1, label="Meus pontos", color="#000000", marker=".", s=400)
plt.legend()
plt.plot(x1, y1, color="#F08080", linestyle="-")
#Método para Exportar figura, salvando nos arquivos da máquina virtual.
plt.savefig("figura_1.png")
```

Alterando Propriedades da Imagem, ex: tamanho, resolução.

```
plt.scatter(x1, y1, label="Meus pontos", color="#000000", marker=".", s=400)
plt.legend()
plt.plot(x1, y1, color="#F08080", linestyle="-")
#Propiedade dpi manipula tamanho e resolução. 300 é uma boa configuração.
plt.savefig("figura_1.png", dpi=300)
```

Formato PDF

```
plt.scatter(x1, y1, label="Meus pontos", color="#000000", marker=".", s=400)
plt.legend()
plt.plot(x1, y1, color="#F08080", linestyle="-")
plt.savefig("figura_1.pdf")
```

# Estudo de Caso
População Brasileira de 1980-2016

DataSus - informações em geral da população brasileira.

**Visualização no formato CSV**

```
dados_pop = open('original.csv').readlines()
dados_pop
```

```
x = []
y = []
```

```
#Percorrer uma lista de valores (dados), ignorando a 1º linha, separando os valores pelo ';' e adicionando os anos ao x e os dados da população ao y.
#Converter o formarto String para Int.
for i in range(len(dados_pop)):
  if i != 0:
    linha = dados_pop[i].split(';')
    x.append(int(linha[0]))
    y.append(int(linha[1]))

print('Anos', x)
print('População', y)
```

Gráfico de Linha

```
plt.plot(x, y)
plt.title("Crescimento da População Brasileira entre 1980 e 2016")
plt.xlabel("Ano")
plt.ylabel("Centena de Milhão")
plt.show()
```

Gráfico de Barra


```
plt.bar(x, y)
plt.title("Crescimento da População Brasileira entre 1980 e 2016")
plt.xlabel("Ano")
plt.ylabel("Centena de Milhão")
plt.show()
```

Combinando Gráficos

```
plt.plot(x, y, color="k", linestyle="--")
plt.bar(x, y, color="#e4e4e4")
plt.title("Gráfico de Barras + Gráfico de Linha")
plt.savefig("pop_brasileira.png", dpi=300)
plt.show()
```

# 2º Estudo de Caso

Human 18S rRNA gene, complete

Base de dados

```
entrada = open('bacteria.fasta').read()
saida = open('bacteria.html', 'w')

cont = {}

for i in ['A', 'C', 'G', 'T']:
  for j in ['A', 'C', 'G', 'T']:
    cont[i+j] = 0

entrada = entrada.replace('\n', "")

for k in range(len(entrada)-1):
  cont[entrada[k]+entrada[k+1]] +=1

saida.write("<div>")

i = 1
for k in cont:
  transparencia = cont[k]/max(cont.values())
  saida.write("<div style= 'width:100px; border: 1px solid #111; height:100px; float:left; background-color:rgba(0, 0, 30, "+str(transparencia)+"');> "+k+" </div>")

  if i%4 == 0:
    saida.write("<div style='clear:both'></div>")

  i+=1

saida.close()
```
