```
#Importando Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Importando famoso dataset Irís e criando Data Frame a partir destes.
df = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
#Visualizando as 5 primeiras linhas do Data Frame.
df.head()
```

```
#Visualizando a estrutura do Data Frame.
df.shape
```

```
#Descrição estatística dos dados do Data Frame.
df.describe()
```

```
#Verificando distribuição de classes.
df['variety'].value_counts()
```

```
df['variety'].value_counts()
```

```
#Removendo a coluna equivalente ao Target dos meus dados e criando um Array somente com as Características.
X = np.array(df.drop('variety', axis=1))
```

```
X
```

```
#Importando algoritmo KMeans da biblioteca Sklearn
from sklearn.cluster import KMeans
```

```
#Instanciando objeto kmeans, definindo o número de clusters e reduzindo a aleatoriedade p/ geração dos dados.
kmeans = KMeans(n_clusters=3, random_state=0)
```

```
#Treinando o Modelo.
kmeans.fit(X)
```

```
#Verificando os resultados de clusterização do Modelo.
kmeans.labels_
```

```
#Inserindo esta nova classe no Data Frame Original p/ comparação.
df['Kclasses'] = kmeans.labels_
```

```
df
```

```
#Visualizando a distribuição de correlação entre os dados originais. 
sns.pairplot(df, hue='variety')
```

```
#Visualizando a distribuição de correlação entre os novos dados. 
sns.pairplot(df, hue='Kclasses')
```
